```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
```

```python
x_a = np.arange(0, np.pi + 0.0001, 0.0001)
y_sin = np.sin(x_a)
sin_av = np.full_like(x_a, 2 / np.pi)
```

```python
plt.plot(x_a, y_sin, label=r"$\sin(t)$")
plt.plot(x_a, sin_av, label=r"$f_{\text{AV}}$")
plt.title(r"$\sin(t)$ vs. $\sin(t)$'s average value over $[0, \pi]$")
plt.xlabel(r"$t$")
plt.ylabel(r"$y$")
plt.legend(loc=r"upper left")
```

```python
x_b = np.arange(1, 2 + 0.0001, 0.0001)
x_t_sqr = x_b**2
t_sqr_av = np.full_like(x_b, 7/3)
```

```python
plt.plot(x_b, x_t_sqr, label=r"$t^2$")
plt.plot(x_b, t_sqr_av, label=r"$f_{\text{AV}}$")
plt.title(r"$t^2$ vs. $t^2$'s average value over $[1, 2]$")
plt.xlabel(r"$t$")
plt.ylabel(r"$y$")
plt.legend(loc=r"upper left")
```

```python
r_h = 1/2
o_h = 2.0 * np.pi
def g(t):
    return np.cos(o_h* t)

def g_bar(t):
    return r_h**-1 * (o_h**-1 * (np.sin(o_h * t) - np.sin(o_h * (t - r_h))))

x_h = np.arange(0, 4 + 0.0001, 0.0001)
y_h_g = g(x_h)
y_h_g_bar = g_bar(x_h)
```

```python
plt.plot(x_h, y_h_g, label=r"$g(t) = \cos(2\pi t)$")
plt.plot(x_h, y_h_g_bar, label=r"$\bar g(t)$")
plt.title(r"$\cos(2\pi t)$ vs. its $\frac{1}{2}$ unit moving average value over $[0, 4]$")
plt.xlabel(r"$t$")
plt.ylabel(r"$y$")
plt.legend(loc=r"upper left")
```
