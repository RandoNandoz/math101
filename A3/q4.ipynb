```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from math import pi

plt.rcParams["figure.dpi"] = 300
```

```python
values = pd.read_csv(r"./A17.csv", names=[r"f"])
```

```python
n = values.iloc[-1].name
```

```python
dx = pi/n
```

```python
x_axis = np.arange(0, np.pi + dx, dx)
```

```python
values['x'] = x_axis
```

```python
values.plot(x='x', y=r"f", title=r"$f(x_i) vs. x_i$", ylabel=r"$y$", xlabel="$x$")
```
