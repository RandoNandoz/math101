```python
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 300
from sympy import *
init_printing()
```

```python
n = Symbol('n', integer=True, positive=True)
```

```python
x = Symbol('x', real=True, positive=True)
```

```python
f = 1
```

```python
B_n_f = (2 / pi) * integrate(
    f * sin(n * x), (x, 0, pi)
)
```

```python
S_f_3 = Sum(
    B_n_f * sin(n * x), (n, 1, 3)
)
```

```python
S_f_5 = Sum(
    B_n_f * sin(n * x), (n, 1, 5)
)
```

```python
S_f_11 = Sum(
    B_n_f * sin(n * x), (n, 1, 11)
)
```

```python
plot(S_f_3, (x, 0, pi), title=r"$S_{N}$ for $N = 3$", legend=True, label=r"$S_3$", line_color='red')
```

```python
plot(S_f_5, (x, 0, pi), title=r"$S_{N}$ for $N = 5$", legend=True, label=r"$S_5$", line_color='green')
```

```python
plot(S_f_11, (x, 0, pi), title=r"$S_{N}$ for $N = 11$", legend=True, label=r"$S_{11}$", line_color='blue')
```

```python
g = x
```
