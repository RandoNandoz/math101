```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import sympy as sp
import pandas as pd
plt.rcParams['figure.dpi'] = 300
```

```python
def f(t):
    for n in range(-100, 100, 1):
        if float.is_integer(t):
            return float('nan')
        elif (2 * n <= t) and (t < 2 * n + 1):
            return 1
        elif (2 * n + 1 <= t) and (t < 2 * n + 2):
            return 0
    print(f"Error on t={t}!")
    
```

```python
dx = 1/10000
values = pd.DataFrame(
    np.arange(
        0, 8 + dx, dx
    ), columns=['x']
)
```

```python
values
```

```python
values['y'] = values['x'].map(
    f
)
```

```python
values['y']
```

```python
plt.axis("equal")

plt.plot(
    values['x'],
    values['y']
)

plt.axis
plt.xlabel(r"$t$")
plt.ylabel(r"$y$")

plt.title(r"$y = f(t)$")

for i in range(0, 9, 1):
    plt.gca().add_patch(
        patches.Circle(
            (i, 1) if i % 2 == 0 else (i, 0),
            radius=0.04
        )            
    )
    
for i in range(0, 9, 1):
    plt.gca().add_patch(
        patches.Circle(
            (i, 0) if i % 2 == 0 else (i, 1),
            radius=0.04, fill=False, color="tab:blue"
        )            
    )

plt.show()
```
